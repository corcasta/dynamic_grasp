```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("standard_behaviour.csv")
df = df.rename(columns={"Unnamed: 0": "Timestep", "0": "Force (Unitless)"})
df.head()
```

```python
df["Force (Unitless)"].plot(title="Expected behaviour ", xlabel="Timestep", ylabel="Force (Unitless)", grid=True)
```

```python
df["Filtered Force (Unitless)"] = df["Force (Unitless)"].rolling(20).mean()
df["STD Force (Unitless)"] = df["Force (Unitless)"].rolling(20).std()
```

```python
df[["Force (Unitless)", "Filtered Force (Unitless)"]].plot(title="Expected behaviour ", xlabel="Timestep", ylabel="Force (Unitless)", grid=True)
```

```python
sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'lightgreen'})

timestep = df["Timestep"]
gross_force = df["Force (Unitless)"] 
filtered_force = df["Filtered Force (Unitless)"]
std = df["STD Force (Unitless)"]

plt.plot(timestep, gross_force, 'g--', label='Gross')
plt.plot(timestep, filtered_force, 'r-', label='Filtered')
plt.fill_between(timestep, filtered_force - std, filtered_force + std, color='r', alpha=0.3)
plt.grid(visible=True, linewidth=1)
plt.xlabel("Timestep")
plt.ylabel("Force (Unitless)")
plt.legend(title='Force')


#plt.show()
```
